### Check null value count in every column

```python
def check_empty_values(df, df_name=None):
    """
    Check for empty/null values in a dataframe and provide detailed information.
    
    Parameters:
    df: pandas DataFrame to check for empty values
    df_name: optional name for the dataframe (for display purposes)
    """
    name = df_name if df_name else "DataFrame"
    
    
    
    print(f"{name}:")
    
    # Get null counts for each column
    null_counts = df.isnull().sum()
    total_nulls = null_counts.sum()
    
    print(f"  Total rows: {len(df)}, Total columns: {len(df.columns)}, Total null values: {total_nulls}")
    
    # Show columns with null values
    columns_with_nulls = null_counts[null_counts > 0]
    all_cols = null_counts[null_counts >= 0]
    non_null_cols = null_counts[null_counts == 0]
    if len(columns_with_nulls) > 0:
        for col, count in all_cols.items():
            percentage = (count / len(df)) * 100
            print(f"    {col}: {count} nulls ({percentage:.1f}%)")
        print(f" {len(non_null_cols)} columns with no nulls: {list(non_null_cols.index)}")
    else:
        print("  No null values found.")
    print("=" * 50)
    
    return non_null_cols
    
# Check empty values for all dataframes
non_null_cols_total = {}
for key, df in dfs.items():
    non_null_cols = check_empty_values(df, key)
    non_null_cols_total[key] = non_null_cols
```

```python
def check_filled_columns(df, required_columns, df_name=None):
    """
    Check if the specified columns in the dataframe are completely filled (no nulls).
    
    Parameters:
    df: pandas DataFrame to check
    required_columns: list of column names that must be completely filled
    df_name: optional name for the dataframe (for display purposes)
    """
    name = df_name if df_name else "DataFrame"
    print(f"Checking required columns in {name}:")
    
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        print(f"  Missing columns: {missing_columns}")
        return False
    
    unfilled_columns = []
    for col in required_columns:
        if df[col].isnull().any():
            unfilled_columns.append(col)
    
    if unfilled_columns:
        print(f"  Columns with missing values: {unfilled_columns}")
        return False
    else:
        print("  All required columns are completely filled.")
        return True

for key, df in dfs.items():
    check_filled_columns(df, filled_cols, df_name=key)
```

## Duplicates

### Duplicate rows

```python
def check_duplicates_rows(df, df_name=None, cols="all"):
    """
    Check for duplicate rows in a dataframe and provide detailed information.
    
    Parameters:
    df: pandas DataFrame to check for duplicates
    df_name: optional name for the dataframe (for display purposes)
    
    Returns:
    dict with duplicate information
    """
    name = df_name if df_name else "DataFrame"
    if cols == "all":
        cols = df.columns.tolist()
        
    # Check for duplicates
    duplicates = df[cols].duplicated()
    num_duplicates = duplicates.sum()
    
    # Get duplicate rows
    duplicate_rows = df[duplicates]
    
    # Check for duplicates keeping all occurrences
    all_duplicates = df[df.duplicated(keep=False)]
    
    print(f"{name}:")
    print(f"  Duplicate rows: {num_duplicates}/{len(df)}")
    # print(f"  Unique rows: {len(df) - num_duplicates}")
    
    if num_duplicates > 0:
        print(f"  Rows involved in duplicates (including originals): {len(all_duplicates)}")
        print(f"  Duplicate row indices: {duplicate_rows.index.tolist()}")
        
        # Show first few duplicate rows
        if len(duplicate_rows) > 0:
            print(f"  First duplicate row(s):")
            print(duplicate_rows.head())
    
    print("=" * 50)
    
    # return {
    #     'total_rows': len(df),
    #     'duplicate_count': num_duplicates,
    #     'unique_count': len(df) - num_duplicates,
    #     'duplicate_indices': duplicate_rows.index.tolist(),
    #     'duplicate_rows': duplicate_rows
    # }

# Check duplicates for all dataframes
for key, df in dfs.items():
    check_duplicates_rows(df, key, cols="all")
```
